```python
### Scraping detailed pet descriptions
## Function to make staggered URL calls to pull in full pet descriptions

def weget_pf(url, file):
    inputs = 'Pet_Profiles_STL/' + file + " " + url
    !wget --wait 2 --random-wait -O {inputs}
    timeDelay = random.random()
    time.sleep(timeDelay)
```

```python
count = 0

for peturl in dog_data_STL_19_full['url']:
    url = peturl
    file = re.search('www.petfinder.com/dog/(.*?)\/', peturl)[1]

    weget_pf(url, file)

    #Give status on number of files.
    if (count % 500 == 0):
        print("500 more have been collected.")

    count += 1
```
